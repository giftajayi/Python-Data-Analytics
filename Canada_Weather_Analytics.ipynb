# **Data Cleaning**

```python
data['Province'] = data['Province'].str.strip()  # Remove extra spaces
data['Province'] = data['Province'].str.upper()  # Convert to uppercase for matching
```

```python
(data['Province'].unique()) # Check unique values in the province column
```

```python
data['Province'] = data['Province'].replace({  #replacing the values in the province column
    'SASK': 'Saskatchewan',
    'BC': 'British Columbia',
    'QUE': 'Quebec',
    'NU': 'Nunavut',
    'PEI': 'Prince Edward Island',
    'NB': 'New Brunswick',
    'MAN': 'Manitoba',
    'ONT': 'Ontario',
    'ALTA': 'Alberta',
    'NS': 'Nova Scotia',
    'NFLD': 'Newfoundland and Labrador'
})
```

```python
(data['Province'].unique()) # Checking to confirm the replacements made
```

```python
data.duplicated().sum() # checking for duplicate values
```

```python
data.drop_duplicates(inplace=True) # drop the duplicate values
```

```python
data.duplicated().sum() # checking for duplicate values to confirm the dropped values
```

```python
data.isnull().sum() # checking for missing values
```

```python
data.shape
```

```python
# Fill missing numerical values with the median (recommended for skewed data)
data['MinTemp'] = data['MinTemp'].fillna(data['MinTemp'].median())
data['MeanTemp'] = data['MeanTemp'].fillna(data['MeanTemp'].median())
data['MaxTemp'] = data['MaxTemp'].fillna(data['MaxTemp'].median())

# Fill missing Latitude and Longitude with their median values
data['Latitude'] = data['Latitude'].fillna(data['Latitude'].median())
data['Longitude'] = data['Longitude'].fillna(data['Longitude'].median())
```

```python
data.isnull().sum() # checking for missing values
```

```python
data.shape
```

# **Data Analysis**

```python
# Group and aggregate
grouped_data = data.groupby('Province').mean()

# Correlation
# Select only numerical columns before calculating correlation
numerical_data = data.select_dtypes(include=np.number)
correlation_matrix = numerical_data.corr()
correlation_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
data['numeric_column'].hist(bins=20)
plt.show()

# Box plots
sns.boxplot(x='category_column', y='numeric_column', data=data)
plt.show()

# Heatmap for correlations
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot bar plot using Seaborn for average 'MaxTemp' by 'Province'
sns.barplot(x='Province', y='MaxTemp', data=data, ci=None)  # `ci=None` to remove confidence intervals
plt.title('Average MaxTemp by Province')
plt.xlabel('Province')
plt.ylabel('Average Max Temperature')
plt.xticks(rotation=90)  # Rotate x-axis labels if they are long
plt.show()
```

```python
# Plot bar plot using Seaborn for average 'MinTemp' by 'Province'
sns.barplot(x='Province', y='MinTemp', data=data, ci=None)  # `ci=None` to remove confidence intervals
plt.title('Average MinTemp by Province')
plt.xlabel('Province')
plt.ylabel('Average Min Temperature')
plt.xticks(rotation=90)
```

```python
# Plot bar plot using Seaborn for average 'MeanTemp' by 'Province'
sns.barplot(x='Province', y='MeanTemp', data=data, ci=None)  # `ci=None` to remove confidence intervals
plt.title('Average MeanTemp by Province')
plt.xlabel('Province')
plt.ylabel('Average Mean Temperature')
plt.xticks(rotation=90)  # Rotate x-axis labels if they are long
plt.show()
```

```python
# Group by 'Year' and calculate the average of 'MinTemp' for each year
min_temp_by_year = data.groupby('Year')['MinTemp'].mean().reset_index()

# Plot the minimum temperature over the years
plt.figure(figsize=(10, 6))

# Plot MinTemp
sns.lineplot(x='Year', y='MinTemp', data=min_temp_by_year, marker='o', color='blue')

# Adding labels and title
plt.title('Average Min Temperature Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Min Temperature')
plt.grid(True)
plt.show()
```

```python
# Group by 'Year' and calculate the maximum of 'MaxTemp' for each year
max_temp_by_year = data.groupby('Year')['MaxTemp'].max().reset_index()

# Plot the maximum temperature over the years
plt.figure(figsize=(10, 6))

# Plot MaxTemp
sns.lineplot(x='Year', y='MaxTemp', data=max_temp_by_year, marker='o', color='red')

# Adding labels and title
plt.title('Maximum Temperature Over the Years')
plt.xlabel('Year')
plt.ylabel('Maximum Temperature')
plt.grid(True)
plt.show()
```

```python
# Group by 'Province' and calculate the average 'MinTemp'
avg_min_temp_by_province = data.groupby('Province')['MinTemp'].mean().reset_index()

# Plot the average minimum temperature by Province
plt.figure(figsize=(12, 6))

# Creating a bar plot
sns.barplot(x='Province', y='MinTemp', data=avg_min_temp_by_province, palette='viridis')

# Adding labels and title
plt.title('Average Min Temperature by Province')
plt.xlabel('Province')
plt.ylabel('Average Min Temperature')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group by 'Province' and calculate the average 'MaxTemp'
avg_max_temp_by_province = data.groupby('Province')['MaxTemp'].mean().reset_index()

# Plot the average maximum temperature by Province
plt.figure(figsize=(12, 6))

# Creating a bar plot
sns.barplot(x='Province', y='MaxTemp', data=avg_max_temp_by_province, palette='viridis')

# Adding labels and title
plt.title('Average Max Temperature by Province')
plt.xlabel('Province')
plt.ylabel('Average Max Temperature')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group by 'Province' and calculate the maximum 'MaxTemp' for each province
max_temp_by_province = data.groupby('Province')['MaxTemp'].max().reset_index()

# Find the province with the highest maximum temperature
highest_temp_province = max_temp_by_province.loc[max_temp_by_province['MaxTemp'].idxmax()]

# Print the result
print(f"The province with the highest temperature is {highest_temp_province['Province']} with a temperature of {highest_temp_province['MaxTemp']}Â°C.")

# Plot the maximum temperature by province
plt.figure(figsize=(12, 6))

# Create the bar plot
sns.barplot(x='Province', y='MaxTemp', data=max_temp_by_province, palette='coolwarm')

# Adding labels and title
plt.title('Maximum Temperature by Province')
plt.xlabel('Province')
plt.ylabel('Maximum Temperature')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
